```python
ind_mktcap = ind_nfirms * ind_size
ind_mktcap.shape
```

```python
total_mktcap = ind_mktcap.sum(axis="columns")
total_mktcap.head()
```

```python
total_mktcap.plot()
```

```python
ind_capweight = ind_mktcap.divide(total_mktcap, axis="rows")
ind_capweight.head()
```

```python
ind_capweight["1926"].sum(axis="columns")
```

```python
ind_capweight[["Fin", "Steel"]].plot(figsize=(12,6))
```

```python
total_market_return = (ind_capweight*ind_return).sum(axis="columns")
total_market_return.plot()
```

```python
total_market_index = rk.drawdown(total_market_return).Wealth
total_market_index.plot(figsize=(10, 5), title="Total Market CapWeighted Index 1926-2018")
```

```python
total_market_index["1980":].plot(figsize=(10,5))
total_market_index["1980":].rolling(window=36).mean().plot(figsize=(10,5))
```

```python
tmi_tr36rets = total_market_return.rolling(window=36).aggregate(rk.annualized_rets, periods_per_year=12)
tmi_tr36rets.plot(figsize=(12,6), label="Trailing 36 Months Return", legend=True)
total_market_return.plot(label="Returns", legend=True)
```

# Rolling Correlation - along with Multi-indices and .groupby

```python
ts_corr = ind_return.rolling(window=36).corr()
ts_corr.tail()
```

```python
ts_corr.index.names = ['date','industry']
ind_tr36corr = ts_corr.groupby(level="date").apply(lambda cormat: cormat.values.mean())
ind_tr36corr.plot()
```

```python
tmi_tr36rets["2007":].plot(label="Trailing 36 Months Returns", figsize=(12,6), legend=True)
ind_tr36corr["2007":].plot(label="Trailing 36 Months Correlation", figsize=(12,6), legend=True, secondary_y=True)
```

```python
# Correlation between returns and correlation: Why diversification fails
tmi_tr36rets.corr(ind_tr36corr)
```
