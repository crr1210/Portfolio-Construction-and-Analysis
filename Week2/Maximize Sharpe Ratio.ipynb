# Maximize Sharpe Ratio and Capital Market Line

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import risk_kit as rk

ind = rk.get_ind_returns()
er = rk.annualized_rets(ind["1996":"2000"], 12)
cov = ind["1996":"2000"].cov()
```

```python
ax = rk.plot_ef(20, er, cov)
ax.set_xlim(left = 0)
rf = 0.1
w_msr = rk.msr(rf, er, cov)
r_msr = rk.portfolio_return(w_msr, er)
vol_msr = rk.portfolio_vol(w_msr, cov)
# Add Capital Market Line(the green line)
cml_x = [0, vol_msr]
cml_y = [rf, r_msr]
ax.plot(cml_x, cml_y, color="green", marker="o", linestyle="dashed")
```
